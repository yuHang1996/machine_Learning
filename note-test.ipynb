#matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
x = np.linspace(0, 10, 100)
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
plt.plot(x, siny, linestyle='--', color='r', label='sin(x)')
# plt.xlim(-5, 5)
plt.plot(x, cosy)
```

```python
plt.plot(x, siny, linestyle='--', color='#ccc', )
plt.plot(x, cosy)
plt.axis([-1, 11, -2, 2])
plt.xlabel('x label')
plt.ylabel('y label')
```

```python
plt.plot(x, siny, linestyle='--', color='#ccc', label='sin(x)')
plt.plot(x, cosy, label='cos(x)')
plt.axis([-1, 11, -2, 2])
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('sin & cos')
plt.legend()
```
